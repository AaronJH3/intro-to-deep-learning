```python
#My main goal for now is to get an idea on what the data has right now

data = pd.read_csv("./data.csv")
print(data)
```

```python
data = data.drop(columns=[
    'Unnamed: 32',        # Unsure what this column is for so removing it then describing again
    'id' #ID is too large - could cause deviation on analysis.
])
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

 
cols_to_scale = [
    'perimeter_mean',
    'perimeter_se',
    'perimeter_worst',
    'area_mean',
    'area_se',
    'area_worst',
    'radius_mean',
    'radius_se',
    'radius_worst',
    'texture_mean',
    'texture_se',
    'texture_worst'

]

scaled_cols = scaler.fit_transform(data[cols_to_scale])

# printing the normailized columns to make sure they are tight.
scaled_cols
```

```python
# So, we just got back an NDArray, and we need to put these
# columns back into a dataframe.
scaled_data = data.copy(deep=True)

for i, col  in enumerate(cols_to_scale):
    scaled_data[col] = scaled_cols[:, i]

scaled_data.head(5)

#Checking a few Rows to ensure the data is looking solid.
```

```python
# Even though we labeled some columns as "category" we still need to one-hot 
# encode them. Pandas makes this super easy too:
scaled_dummy_df = pd.get_dummies(scaled_data)
scaled_dummy_df.head(1)

# Note that this takes awhile, it's procecssing a lot of data. 
# Note also that pandas automatically looks for columns with
# a categorical type, so being explicit above was important 
# to making this part easy.
```

```python
# Note that all our numeric columns are between 0 and 1, except SALE PRICE
# All that's left to do here is to separate the labels from the features.

x_trainingdata = scaled_dummy_df.drop(columns=['diagnosis_B', 'diagnosis_M'])
 
x_trainingdata.head(1)

print(x_trainingdata)
 
```

```python
drop_list = ['diagnosis_B', 'diagnosis_M']
#drop_list = ['diagnosis_B']

y_traininglabel = scaled_dummy_df.drop(scaled_dummy_df.columns.difference(drop_list), axis=1)

y_traininglabel.head(1)

print(y_traininglabel)
```

```python
def plot_training_history(history, model):
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model accuracy')
    plt.ylabel('accuracy')
    plt.xlabel('epoch')
    plt.legend(['training', 'validation'], loc='best')

    plt.show()

    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['training', 'validation'], loc='best')

    plt.show()
    
    loss, accuracy  = model.evaluate(x_trainingdata, y_traininglabel, verbose=False)
    print(f'Test loss: {loss:.3}')
    print(f'Test accuracy: {accuracy:.3}')
```

```python
# Sweet, lets make a simple neural net with keras to make sure we can run the data
# through it. We don't expect great predictions out of this simple model we just
# want to be sure that we can :
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()


#print(len(x_train.columns))
# Sigmoid and other functions that squash the output might not be
# very appropriate for this task, because our target values are 
# quite large!
model.add(Dense(units=300, activation='relu', input_shape=(len(x_trainingdata.columns),)))

model.add(Dense(units=200, activation='relu'))

model.add(Dense(units=100, activation='relu'))

model.add(Dense(units=10, activation='relu'))

# For regression it's common to use a linear activation function
# since our output could be anything. In our case, it would never
# make sense to guess less than 0, so I'm using relu
model.add(Dense(units=1, activation='relu'))

# This function provides useful text data for our network
model.summary()
```

```python
from tensorflow.keras.layers import LeakyReLU

# MSE is pretty common for regression tasks
model.compile(optimizer="adam", loss='mean_squared_error', metrics=['accuracy'])
history = model.fit(x_trainingdata, y_traininglabel, batch_size=128, epochs=20, verbose=True, validation_split=.1)


plot_training_history(history, model)

```

```python
#==============

model = Sequential()
model.add(Dense(units=10,  input_shape=(len(x_trainingdata.columns),)))
model.add(LeakyReLU(alpha=.01))
 
model.add(Dense(units=10))
model.add(LeakyReLU(alpha=.01))

model.add(Dense(units=2, activation='softmax'))
model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

history = model.fit(x_trainingdata, y_traininglabel, batch_size=128, epochs=20, verbose=False, validation_split=.1)
plot_training_history(history, model)
```
